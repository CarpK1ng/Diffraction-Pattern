```python
#TiAl α2
import math
import numpy as np
a = 5.75
c = 4.63
print("\nlength of g vector:")
d1 = 10 / float(input("|g1|(1/nm): "))
d2 = 10 / float(input("|g2|(1/nm): "))
print("\ninterplanar spacing:")
print("d1 = ", round(d1, 2), "Å")
print("d2 = ", round(d2, 2), "Å\n")
angle = float(input("angle (degree): "))
cosine = math.cos(angle/180*math.pi)
LHS1 = 1/(d1*d1)
LHS2 = 1/(d2*d2)
plane1 = []
plane2 = []

def P12(h1, h2, k1, k2, l1, l2, a, c):
    return ((4*(h1*h2+k1*k2)+2*(h1*k2+h2*k1))/(3*a*a))+((l1*l2)/(c*c))

def Pnn(h, k, l, a, c):
    return ((4*(h*h+h*k+k*k))/(3*a*a)) + (l*l)/(c*c)

for h in range(2,-3,-1):
    for k in range(2,-3,-1):
        for l in range(2,-3,-1):
            RHS = (4/3)*((h*h+h*k+k*k)/(a*a))+((l*l)/(c*c))
            if abs(RHS-LHS1)<0.01:
                anti_parallel = True
                for idx in range(len(plane1)):
                    if plane1[idx]["h"]*(-1) == h and plane1[idx]["k"]*(-1) == k and plane1[idx]["l"]*(-1) == l:
                        anti_parallel = False
                if anti_parallel == True:
                    plane1.append({"h":h, "k":k, "l":l})
for h in range(2,-3,-1):
    for k in range(2,-3,-1):
        for l in range(2,-3,-1):
            RHS = (4/3)*((h*h+h*k+k*k)/(a*a))+((l*l)/(c*c))
            if abs(RHS-LHS2)<0.01:
                anti_parallel = True
                for idx in range(len(plane2)):
                    if plane2[idx]["h"]*(-1) == h and plane2[idx]["k"]*(-1) == k and plane2[idx]["l"]*(-1) == l:
                        anti_parallel = False
                if anti_parallel == True:
                    plane2.append({"h":h, "k":k, "l":l})
print("\nthe planes that match the interplanar spacing d1:")
for _ in plane1:
    print("( h1 k1 l1 ) = ",_["h"],_["k"],_["l"], end="   ")
print("\n\nthe planes that match the interplanar spacing d2:")
for _ in plane2:
    print("( h2 k2 l2 ) = ",_["h"],_["k"],_["l"], end="   ")
print("\n-----------------------------------------------")
for i in range(0, len(plane1)):
    h1 = plane1[i]["h"]
    k1 = plane1[i]["k"]
    l1 = plane1[i]["l"]
    pt = "( h1 k1 l1 ) = "+str(h1)+" "+str(k1)+" "+str(l1)+" \n"
    lens = len(pt)
    for j in range(0, len(plane2)):
        h2 = plane2[j]["h"]
        k2 = plane2[j]["k"]
        l2 = plane2[j]["l"]
        p12 = abs(P12(h1,h2,k1,k2,l1,l2,a,c))
        p1 = math.sqrt(abs(Pnn(h1,k1,l1,a,c)))
        p2 = math.sqrt(abs(Pnn(h2,k2,l2,a,c)))
        zone = np.cross([h1, k1, l1],[h2, k2, l2])

        gcd = math.gcd(math.gcd(abs(zone[0]), abs(zone[1])), abs(zone[2]))
        if gcd > 1:
            zone = zone // gcd

        if abs(((p12)/(p1*p2))-cosine)<0.05:
          pt += "    ( h2 k2 l2 ) = " + str(h2) + " " + str(k2) + " " + str(l2) + "  zone:  [ " + str(int(zone[0])) + " " + str(int(zone[1])) + " " + str(int(zone[2])) + " ]\n"
          #print((p12)/(p1*p2), p12, p1, p2)
    if len(pt) > lens:
      print(pt)
      print("angle = ", round(np.arccos(p12/(p1*p2))/math.pi*180, 2), "°")
```

```python
#General version for the 7 crystal systems
import math
import numpy as np
from sympy import symbols

def cos(x):
    return math.cos(x * math.pi / 180)

crystal_system = float(input("crystal system: \n"
                  "  cubic: 1\n"
                  "  tetragonal: 2\n"
                  "  orthorhombic: 3\n"
                  "  hexagonal: 4\n"
                  "  trigonal: 5\n"
                  "  monoclinic: 6\n"
                  "  triclinic: 7\n" ))
print("\nlattice parameters:")
if crystal_system == 1:
    α = β = γ = 90
    a = float(input("  a(Å): "))
    c = b = a
if crystal_system == 2:
    α = β = γ = 90
    a = float(input("  a(Å): "))
    c = float(input("  c(Å): "))
    b = a
if crystal_system == 3:
    α = β = γ = 90
    a = float(input("  a(Å): "))
    b = float(input("  b(Å): "))
    c = float(input("  c(Å): "))
if crystal_system == 4:
    α = β = 90
    γ = 120
    a = float(input("  a(Å): "))
    c = float(input("  c(Å): "))
    b = a
if crystal_system == 5:
    α = β = 90
    γ = 120
    a = float(input("  a: "))
    c = float(input("  c: "))
    b = a
if crystal_system == 6:
    α = γ = 90
    β = float(input("  β: "))
    a = float(input("  a: "))
    b = float(input("  b: "))
    c = float(input("  c: "))
if crystal_system == 7:
    α = float(input("  α: "))
    β = float(input("  β: "))
    γ = float(input("  γ: "))
    a = float(input("  a: "))
    b = float(input("  b: "))
    c = float(input("  c: "))

metric_tensor = np.linalg.inv(np.array([[a * a, a * b * cos(γ), a * c * cos(β)],
                          [a * b * cos(γ), b * b, b * c * cos(α)],
                          [a * c * cos(β), b * c * cos(α), c * c]]))
print("\nlength of g vector:")
d1 = 10 / float(input("|g1|(1/nm): "))
d2 = 10 / float(input("|g2|(1/nm): "))
print("\ninterplanar spacing:")
print("d1 = ", round(d1, 2), "Å")
print("d2 = ", round(d2, 2), "Å\n")
angle = float(input("angle (degree): "))

h1, k1, l1, h2, k2, l2 = symbols('h1, k1, l1, h2, k2, l2')

P1 = np.array([h1, k1, l1])
P2 = np.array([h2, k2, l2])

# 初始化儲存 Miller index的list
P1_indices = []
P2_indices = []

d1_ = 1 / (((P1.dot(metric_tensor)).dot(P1.T)) ** 0.5)
d2_ = 1 / (((P2.dot(metric_tensor)).dot(P2.T)) ** 0.5)

# 找到符合平面間距的 P1 和 P2 的 Miller index組合
for h in range(2, -3, -1):
    for k in range(2, -3, -1):
        for l in range(2, -3, -1):
            d1_value = d1_.subs({h1: h, k1: k, l1: l}).evalf()  # 替換符號並計算
            if abs(d1_value - d1) < 0.05:
                anti_parallel = True
                for idx in range(len(P1_indices)):
                    if P1_indices[idx]["h"] * (-1) == h and P1_indices[idx]["k"] * (-1) == k and P1_indices[idx]["l"] * (-1) == l:
                        anti_parallel = False
                if anti_parallel:
                    P1_indices.append({"h": h, "k": k, "l": l})

for h in range(2, -3, -1):
    for k in range(2, -3, -1):
        for l in range(2, -3, -1):
            d2_value = d2_.subs({h2: h, k2: k, l2: l}).evalf()  # 替換符號並計算
            if abs(d2_value - d2) < 0.05:
                anti_parallel = True
                for idx in range(len(P2_indices)):
                    if P2_indices[idx]["h"] * (-1) == h and P2_indices[idx]["k"] * (-1) == k and P2_indices[idx]["l"] * (-1) == l:
                        anti_parallel = False
                if anti_parallel:
                    P2_indices.append({"h": h, "k": k, "l": l})

print("\nthe planes that match the interplanar spacing d1:")
for _ in P1_indices:
    print("( h1 k1 l1 ) = ",_["h"],_["k"],_["l"], end="   ")
print("\n\nthe planes that match the interplanar spacing d2:")
for _ in P2_indices:
    print("( h2 k2 l2 ) = ",_["h"],_["k"],_["l"], end="   ")
print("\n---------------------------------------------------------------------------")
# 轉換為具體的數值來計算夾角
for p1 in P1_indices:
    for p2 in P2_indices:
        P1_numeric = np.array([p1["h"], p1["k"], p1["l"]], dtype=int)
        P2_numeric = np.array([p2["h"], p2["k"], p2["l"]], dtype=int)
        angle_ = np.arccos(np.dot(P1_numeric, P2_numeric.T) / (np.linalg.norm(P1_numeric) * np.linalg.norm(P2_numeric)))
        if abs(angle_ / math.pi * 180 - angle) < 0.5:
          zone = np.cross(P1_numeric, P2_numeric)
          gcd = math.gcd(math.gcd(abs(zone[0]), abs(zone[1])), abs(zone[2]))
          if gcd > 1:
            zone = zone // gcd
          print(f"(h1 k1 l1) = {p1['h']} {p1['k']} {p1['l']}, (h2 k2 l2) = {p2['h']} {p2['k']} {p2['l']}, angle = {round(angle_ * 180 / math.pi, 2)}°", f"zone: [ {int(zone[0])} {int(zone[1])} {int(zone[2])} ]\n")
```
